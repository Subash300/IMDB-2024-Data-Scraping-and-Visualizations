```python
                    ## Animation ##
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd
import time
import re

# Start the WebDriver
driver = webdriver.Chrome()
driver.get('https://www.imdb.com/search/title/?title_type=feature&release_date=2024-01-01,2024-12-31&genres=animation')

# Wait for page to load
time.sleep(3)
driver.maximize_window()

# Data lists to store movie information
movie_title = []
movie_duration = []
movie_rating = []
movie_voting = []
movie_genre = "animation"

# Scroll and scrape data
scrolling = True
while scrolling:
    body = driver.find_element(By.TAG_NAME, 'body')
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(2)  # Adding delay to allow content to load

    # Find movie elements
    movies = driver.find_elements(By.XPATH, "//*[@id='__next']/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/ul/li")
    
    try:
        # Click the "See more" button to load more movies
        page_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//span[@class='ipc-see-more__text']")))
        actions = ActionChains(driver)
        actions.move_to_element(page_button).click().perform()
    except:
        # If no more pages are available, stop scrolling
        scrolling = False

    # Scrape movie details
for movie in movies:
        try:
            title = movie.find_element(By.XPATH, './div/div/div/div[1]/div[2]/div[1]/a/h3').text
            duration = movie.find_element(By.XPATH, './div/div/div/div[1]/div[2]/div[2]/span[2]').text
            rating = movie.find_element(By.XPATH, './div/div/div/div[1]/div[2]/span/div/span/span[1]').text
            votes = movie.find_element(By.XPATH, './div/div/div/div[1]/div[2]/span/div/span/span[2]').text
            
            movie_title.append(title)
            movie_duration.append(duration)
            movie_rating.append(rating)
            movie_voting.append(votes)
        except Exception as e:
            print("Error processing movie:", e)

# Function to clean movie name (remove leading digits and period)
def movie_name(name):
    return re.sub(r'^\d+\.', '', name).strip()

# Function to convert duration to minutes
def duration_to_minutes(duration):
    hours = 0
    minutes = 0
    if 'h' in duration:
        hours = int(re.search(r'(\d+)h', duration).group(1))
    if 'm' in duration:
        minutes = int(re.search(r'(\d+)m', duration).group(1))
    return hours * 60 + minutes

# Function to clean vote count (convert 'K' to thousands)
def votes(vote_count):
    vote_count = vote_count.replace('(', '').replace(')', '').strip()
    if 'K' in vote_count:
        return int(float(vote_count.replace('K', '').strip()) * 1000)
    return int(vote_count)

# Clean the data
movie_title = [movie_name(title) for title in movie_title]
movie_duration = [duration_to_minutes(duration) for duration in movie_duration]
movie_voting = [votes(vote) for vote in movie_voting]

# Prepare movie data
movie_data = {
    'Name': movie_title,
    'Genre': movie_genre,
    'Duration': movie_duration,
    'Rating': movie_rating,
    'Vote Count': movie_voting
}

# Create the DataFrame
movies_animation = pd.DataFrame(movie_data)

# Optionally, save the DataFrame to a CSV file
csv_filename = "movies_animation_data.csv"
movies_animation.to_csv("movies_animation_data.csv", index=False)

# Display the result (optional)
#print(movies_list)

# Quit the driver
driver.quit()
```
